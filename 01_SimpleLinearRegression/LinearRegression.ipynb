```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('SampleData.csv')
data
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(data.area,data.price)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(data[['area']],data.price)
```

```python
price1 = reg.predict([[3300]])
print(float(price1))
```

```python
pdata = pd.read_csv('PredictionData.csv')
predictedData= reg.predict(pdata)
print(predictedData)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(data.area,data.price)
plt.plot(data.area,reg.predict(data[['area']]),color='red')
plt.show()
```
